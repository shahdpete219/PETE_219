```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio
import numpy as np
```

```python
las = lasio.read('C:\\Users\\ASUS\\Desktop\\1051661161.LAS')
df = las.df()
print(df.keys())
df['Depth'] = df.index
```

```python
df.describe()
```

```python
df.plot(x='Depth', y='CNSS', c='black', lw=0.5, legend=True, figsize=(8,5))

plt.xlim(3000, 200)
plt.ylim(15,50)
plt.title('Plot Without Any Fill')
plt.show()
```

```python
df.plot(x='Depth', y='CNSS', c='black', lw=0.5, legend=True, figsize=(8,5))

plt.fill_betweenx(df['CNSS'], df['Depth'], 0, facecolor='green')

plt.xlim(3000, 200)
plt.ylim(15,50)
plt.title('Plot Without Any Fill')
plt.show()
```
